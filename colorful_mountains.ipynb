```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
#create figure
f = plt.figure(figsize=(50,10))
#turbo color map
cmap_turbo=cm.get_cmap('turbo')

'''add single subplot to figure
with empty x and y ticklabels and no axes shown'''

ax=f.add_subplot(111,xticks=[],yticks=[])
ax.set_axis_off()

#set the limits to x and y axes
ax.set_xlim([0, 20*np.pi])
ax.set_ylim([0,15.])

#create an x arrange from 0 to 2pi
x = np.linspace(0,20*np.pi,100)

#number of mountain ranges
nmr = 10

#number of peaks in each range
npeak=20



plt.fill_between(x,15,0,zorder=nmr-i,color="#FFFFC5",cmap='turbo')
#loop for each range

for i in range(nmr):
    #set color of range by providing a number from 0 to 1
    mr_color = cmap_turbo(float(i)/float(npeak))
    
    #random set of frequencies
    freq=3.0*np.pi*np.random.rand(npeak)+.1
    
    #random set of phase shifts
    phi=2.0*np.pi*np.random.rand(npeak)+.1
    
    #random amplitude
    amp=1.*np.random.rand(npeak)+.1  
    
    #y offset: each range is offset along y direction
    yoff = 1.*float(i)
    
    #create mountain range
    y = np.zeros_like(x)
    
    #for every peak:
    for j in range (npeak):
        
        w = np.arange(0,np.pi,20)
        #add peak as a sine curve with a random amplitude, frequency, phase
        y += (amp[j])*np.sin(freq[j]*x + phi[j])
        
    #take abs val of sin curve and add the offset
    y = np.fabs(y)+yoff
    plt.fill_between(x,y,0,zorder=nmr-i,color=mr_color,cmap='turbo')
    
plt.savefig('mountain_range.png',bbox_inches='tight',pad_inches=0,dpi=600)
```
