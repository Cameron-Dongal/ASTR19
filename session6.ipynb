```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

take x as parameter and returns sin(x)

```python
def sin(x):
    return np.sin(x)
```

take x as parameter and returns cos(x)

```python
def cos(x):
    return np.cos(x)
```

plot x vs sin(x) and cos(x) over a period from 0 to 2pi

```python

x = np.linspace(0, 2*np.pi, 1000)
y1 = sin(x)
y2 = cos(x)

plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.xlabel("x")
plt.ylabel("sin(x) and cos(x)")
plt.show()
```

print first 10 values of x, sin(x), cos(x)
        

```python
for i in range(10):
    print(x[i], y1[i], y2[i])
```
